```python
from datasets import load_dataset
import pandas as pd

# Using wikiann - pure parquet, no loading script, great NER dataset
print("Loading NER Dataset (wikiann)...")
ner_data = load_dataset("wikiann", "en")
print(ner_data)

# Load AG News for classification
print("\nLoading AG News...")
agnews = load_dataset("ag_news")
print(agnews)
```

```python
# ---- NER Dataset EDA ----
print("=== NER Dataset (WikiANN) ===")
print(f"Train size: {len(ner_data['train'])}")
print(f"Validation size: {len(ner_data['validation'])}")
print(f"Test size: {len(ner_data['test'])}")

# Sample example
print("\nSample NER example:")
sample = ner_data['train'][0]
print(f"Tokens: {sample['tokens']}")
print(f"NER Tags: {sample['ner_tags']}")
print(f"Spans: {sample['spans']}")

# Tag distribution
import collections
all_tags = [tag for example in ner_data['train'] for tag in example['ner_tags']]
tag_counts = collections.Counter(all_tags)
print(f"\nTag distribution: {dict(tag_counts)}")

# ---- AG News EDA ----
print("\n=== AG News Dataset ===")
print(f"Train size: {len(agnews['train'])}")
print(f"Test size: {len(agnews['test'])}")

# Label mapping
label_names = agnews['train'].features['label'].names
print(f"Classes: {label_names}")

# Sample example
print("\nSample AG News example:")
sample2 = agnews['train'][0]
print(f"Text: {sample2['text'][:200]}")
print(f"Label: {label_names[sample2['label']]}")

# Class distribution
labels = [ex['label'] for ex in agnews['train']]
label_counts = collections.Counter(labels)
print(f"\nClass distribution: {dict(label_counts)}")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# NER tag distribution
tag_names = {0:'O', 1:'B-PER', 2:'I-PER', 3:'B-ORG', 4:'I-ORG', 5:'B-LOC', 6:'I-LOC'}
tags = [tag_names[k] for k in sorted(tag_counts.keys())]
counts = [tag_counts[k] for k in sorted(tag_counts.keys())]
axes[0].bar(tags, counts, color='steelblue')
axes[0].set_title('NER Tag Distribution (WikiANN Train)')
axes[0].set_xlabel('Tag')
axes[0].set_ylabel('Count')

# AG News class distribution
class_names = [label_names[k] for k in sorted(label_counts.keys())]
class_counts = [label_counts[k] for k in sorted(label_counts.keys())]
axes[1].bar(class_names, class_counts, color='coral')
axes[1].set_title('AG News Class Distribution')
axes[1].set_xlabel('Class')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('../data/eda_plots.png')
plt.show()
print("EDA plot saved!")
```
